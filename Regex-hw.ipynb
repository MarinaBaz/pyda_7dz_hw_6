## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером. Если да, то функция должна возвращать отдельно номер и регион.

```python
def valid_transport_numbe(str_avto):
    """
    Функция принимает на вход строку и проверяет является ли она валидным транспортным номером вида:
    буква 3 цифры 2 буквы 2 или 3 фифры
    и возвращает отдельно номер и регион.
    
    """
    import re
    pattern = '([АВЕКМНОРСТУХ]{1}\d{3}[АВЕКМНОРСТУХ]{2})(\d{2,3})'
    result = re.match(pattern, str_avto)
    if result == None:
        print(f'транспортный номер {str_avto} не валидный!')
    else:
        list_result = dict()
        list_result['номер'] = result.group(1)
        list_result['регион'] = result.group(2)
        return list_result

#my_avto = 'К7777КТ43'
#my_avto = 'К750КТ43'
my_avto = 'К750КТ750'
print(valid_transport_numbe(my_avto))
```

## Задание 2

Напишите функцию, которая проверяет e-mail адрес на валидность по стандарту RFC822: http://www.w3.org/Protocols/rfc822/. Для этого постарайтесь написать (в крайнем случае найти) самый полный шаблон для проверки e-mail адресов через регулярное выражение. Разберитесь и напишите примеры, какие ваш шаблон может не охватить.

```python
def valid_email(str_email):
    import re
    str_mes = ''
    pattern = '([a-z0-9_-]+\.)*[a-z0-9_-]+@[a-z0-9-]+(\.[a-z0-9-]+)*\.[a-z]{2,6}'
    result = re.match(pattern, str_email)
    if result == None:
        str_mes = f'email: {str_email} не прошел проверку!'
    else:
        str_mes = f'email: {str_email} верный!'
    return str_mes

my_email = 'mpupysheva@gmail.com'
print(valid_email(my_email))
```

**Разберем выражение по частям:**
	
###### ([a-z0-9_-]+\.)*
позволяет ввести:
* в [] указан допустимый диапазон символов: символы латинского алфавита от a до z в нижнем регистре, цифры и символы ‘_’ и ‘-’.
* квантор ‘+’ после [] означает, что должен быть введен хотя бы один символ из диапазона, указанного в [].
* в выражении ‘\.’ ‘\’ является экранирующим символом, и указывает на то, что ‘.’ является не диапазоном любых символов, а метасимволом (т.е. в конце выражения должна присутствовать точка).
* ‘*’ в ‘()*’ означает, что все, что находится в () может повторяться 0 или более раз.
**Пример: ivan19.ivanovich.**

###### [a-z0-9_-]+@[a-z0-9-]+
позволяет ввести:
* символы латинского алфавита от a до z в нижнем регистре, цифры и символы ‘_’ и ‘-’.
* квантор ‘+’ после [] означает, что должен быть введен хотя бы один символ из диапазона, указанного в [].
* символ ‘@’.
* символы латинского алфавита от a до z в нижнем регистре, цифры и символ ‘-’.
* квантор ‘+’ после [] означает, что должен быть введен хотя бы один символ из диапазона, указанного в [].
**Пример: ivanov2000@example_2000**

###### (\.[a-z0-9-]+)*
позволяет ввести:
* в выражении ‘\.’ ‘\’ является экранирующим символом, и указывает на то, что ‘.’ является не диапазоном любых символов, а метасимволом (т.е. в начале выражения должна присутствовать точка).
* символы латинского алфавита от a до z в нижнем регистре, цифры и символ ‘-’.
* квантор ‘+’ после [] означает, что должен быть введен хотя бы один символ из диапазона, указанного в [].
* ‘*’ в ‘()*’ означает, что все, что находится в () может повторяться 0 или более раз.
**Пример: .example**

###### \.[a-z]{2,6} 
позволяет ввести:
* в выражении ‘\.’ ‘\’ является экранирующим символом, и указывает на то, что ‘.’ является не диапазоном любых символов, а метасимволом (т.е. в начале выражения должна присутствовать точка).
* символы латинского алфавита.
* в кванторе {} указано допустимое количество символов, в данном случае – от 2 до 6.
**Пример: .com**

Данный шаблон позволит проверить, например, следующие e-mail адреса:
ivanov@example.com;
ivan_ivanov@example.com;
ivan19.ivanovich.ivanov2000@example_2000.example.com.

Для добавления возможности ввода русскоязычного почтового ящика можно к «a-z» добавить «а-я», регулярное выражение будет выглядеть так:
([a-zа-я0-9_-]+\.)*[a-zа-я0-9_-]+@[a-zа-я0-9-]+(\.[a-zа-я0-9-]+)*\.[a-zа-я]{2,6}

## Задание 3
Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки.

```python
def one_word(str_text):
    import re
    str_mes = ''
    str_res = ''
    result = re.findall('[\w]+', str_text)
    if result == None:
        str_mes = 'в тексте нет последовательных повторов слов!'
    else:
        list_one = list()
        [list_one.append(el) for el in result if el not in list_one]
        for x in list_one:
            str_res += ' ' + x
        str_res = '\n' + str_res.strip()
        str_mes = f'из текста удалены все последовательные повторы слов: {str_res}'
    return str_mes


my_text = 'удалить все последовательные повторы повторы повторы повторы слов слов из заданной строки'
print(one_word(my_text))
```
